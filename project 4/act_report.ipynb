### Most dog were rated between 9 - 13.

```python
plt.figure(figsize= (10,5))
sns.countplot(data= tweet_merge, x= 'rating', color = 'g').set(title = 'Frequency Plot of Dog Rating')
```

The plot show that most dogs were rated between 9 - 13 while the mode rating was 12.

### iPhone is the major source of the pictures

```python
plt.figure(figsize= (7,5))
sns.countplot(data= tweet_merge, x= 'source', color = 'b').set(title = 'Frequency plot of Source')
```

Over 1600 of the the dog picture posted were from iPhone making up for over 95% of the total. This means a very small percentage of picture came from the other sources which are web client and tweetdeck

### Favorite count has a strong positive linear relationship with retweet count

```python
plt.figure(figsize = (10, 7))
sns.scatterplot(data= pl_df , x = 'retweet_count', y = 'favorite_count', hue = 'rating').set(title = 'Retweet count vs Favorite count')
```

The scatterplot of favorite count and retweet count along with the rating converges at the bottom left with a narrow spread as it moves towards the top left. This show a strong positive linear relationship between favorite count and retweet count as expected which means the more like a dog picture get the more the retweet and vice versa. With more thick colors at the top it is obvious that dog with higher rating are more likely going to get more like and retweet.

### So many Retrievers

```python
plt.figure(figsize = (10, 5))
sns.countplot(data= tweet_merge, y = 'pre_breed', order = tweet_merge.pre_breed.value_counts().iloc[:10].index, color = 'b').set(title = 'Top predicted dog breeds')
```

```python
plt.figure(figsize = (12, 6))
sns.boxplot(y = 'pre_breed', x = 'rating',
                    data = pl_df,
                    orient = 'h',
                    width = 0.5,
                    palette = "colorblind",
                    order = pl_df.pre_breed.value_counts().iloc[:10].index).set(title = 'Rating Distribution of Top 10 predicted')
```

The top 10 dog breeds predicted by the neural network shows more of golden retrievers and labrador retrievers. The top 10 breeds had a rating distribution similar to the entire data with more rating around 10 to 12
